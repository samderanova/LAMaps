```python
import osmnx as ox
import folium
import numpy as np
from pathlib import Path
import sys

from dotenv import load_dotenv
load_dotenv()

from src.matrix import fit_to_map, scale_and_place, get_map, get_distance_miles
```

```python
G = get_map()
```

```python
coordinates = np.array([[0, 0], [200, 0], [200, 200], [0, 200]])
starting_point = [35.0549, -118.2426]
```

```python
matrix = scale_and_place(coordinates, starting_point)
result, loss_curve = fit_to_map(matrix)
print(result)

nodes = G.nodes()

closest = []
for r in result:
    node_id = ox.nearest_nodes(G, r[0], r[1])
    node = nodes[node_id]
    closest.append([node['y'], node['x']])

print(closest)
```

```python
m = folium.Map(starting_point, zoom_start=12)

for r in result:
    folium.Marker(
        location=r,
        tooltip="pred",
        popup="pred",
        icon=folium.Icon(icon="cloud"),
    ).add_to(m)

for c in closest:
    folium.Marker(
        location=c,
        tooltip="acc",
        popup="acc",
        icon=folium.Icon(icon="thumb"),
    ).add_to(m)

m
```
