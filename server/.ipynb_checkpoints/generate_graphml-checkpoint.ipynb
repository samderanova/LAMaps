```python
import osmnx as ox
import numpy as np
```

```python
G = ox.graph_from_place("Los Angeles, California, USA", network_type="drive")
print(len(G))
```

```python
ox.plot_graph(G)
```

```python
G_proj = ox.project_graph(G)
G2 = ox.consolidate_intersections(G_proj, rebuild_graph=True, tolerance=30, dead_ends=False)
len(G2)
```

```python
ox.plot_graph(G2)
```

```python
ox.save_graphml(G2, "./LA.graphml")
```
